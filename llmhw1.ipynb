```python
import textwrap
#import lmppl
import torch
import math
import gc
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
def count_words(text):
  words = text.split()
  return len(words)
```

```python

tokenizer1 = AutoTokenizer.from_pretrained("bigscience/bloom-560m")
model1 = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")
inputs1 = tokenizer1.encode("Once upon a time ", return_tensors="pt")
outputs1 = model1.generate(inputs1, min_length=230, max_length=235, do_sample=True,num_beams=6, top_k = 4, no_repeat_ngram_size=2)
#print(tokenizer1.decode(outputs1[0]))
#print(count_words(tokenizer1.decode(outputs1[0])))

wrapped_text= textwrap.fill(tokenizer1.decode(outputs1[0], skip_special_tokens=True, width=100))
# Print the story
print(wrapped_text)
print(count_words(wrapped_text))

#find ttr
types = len(set(wrapped_text))
count = count_words(wrapped_text)
ttr1 = types/count
print(f"The type token ratio of the text is {ttr1:.2f}")

#perplexity
# Tokenize the text and add special tokens
input_ids = tokenizer1(wrapped_text, return_tensors="pt").input_ids
# Get the model output
output = model1(input_ids, labels=input_ids)
# Get the loss value
loss = output.loss
# Calculate the perplexity as the exponent of the loss
ppl1 = torch.exp(loss).item()
# Print the result
print(f"Perplexity of the text is {ppl1:.2f}")
```

```python
tokenizer6 = AutoTokenizer.from_pretrained("bigscience/bloomz-1b7")
model6 = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7")
inputs6 = tokenizer6.encode("Once upon a time ", return_tensors="pt")
outputs6 = model6.generate(inputs6, min_length=230, max_length=235, do_sample=True, num_beams=6, top_k = 4, no_repeat_ngram_size=2,
                          top_p=0.9, temperature = 0.8, repetition_penalty=1)

wrapped_text= textwrap.fill(tokenizer6.decode(outputs6[0], skip_special_tokens=True, width=100))
# Print the story
print(wrapped_text)
print(count_words(wrapped_text))

#find ttr
types = len(set(wrapped_text))
count = count_words(wrapped_text)
ttr6 = types/count
print(f"The type token ratio of the text is {ttr6:.2f}")

#perplexity
input_ids = tokenizer6(wrapped_text, return_tensors="pt").input_ids
# Get the model output
output = model6(input_ids, labels=input_ids)
# Get the loss value
loss = output.loss
# Calculate the perplexity as the exponent of the loss
ppl6 = torch.exp(loss).item()
# Print the result
print(f"Perplexity of the text is {ppl6:.2f}")
```
